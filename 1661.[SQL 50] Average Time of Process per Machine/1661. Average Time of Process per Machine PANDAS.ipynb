```python
import pandas as pd
```

```python
data = [[0, 0, 'start', 0.712], [0, 0, 'end', 1.52], [0, 1, 'start', 3.14], [0, 1, 'end', 4.12], [1, 0, 'start', 0.55], [1, 0, 'end', 1.55], [1, 1, 'start', 0.43], [1, 1, 'end', 1.42], [2, 0, 'start', 4.1], [2, 0, 'end', 4.512], [2, 1, 'start', 2.5], [2, 1, 'end', 5]]
activity = pd.DataFrame(data, columns=['machine_id', 'process_id', 'activity_type', 'timestamp']).astype({'machine_id':'Int64', 'process_id':'Int64', 'activity_type':'object', 'timestamp':'Float64'})
activity
```

```python
max_df = activity[activity['activity_type']=='start'].groupby(['machine_id','process_id']).agg( start_timestamp = ('timestamp','max')).reset_index()
min_df = activity[activity['activity_type']=='end'].groupby(['machine_id','process_id']).agg( end_timestamp = ('timestamp','max')).reset_index()

merged = pd.merge(max_df, min_df, on = ['machine_id','process_id'], how = 'inner')
merged['processing'] = merged.end_timestamp - merged.start_timestamp

merged.groupby('machine_id').agg(
    processing_time = ('processing','mean')
).reset_index().round(3)
```
