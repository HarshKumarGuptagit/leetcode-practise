```python
import pandas as pd
```

```python
data = [[1, '2015-01-01', 10], [2, '2015-01-02', 25], [3, '2015-01-03', 20], [4, '2015-01-04', 30]]
weather = pd.DataFrame(data, columns=['id', 'recordDate', 'temperature']).astype({'id':'Int64', 'recordDate':'datetime64[ns]', 'temperature':'Int64'})
```

```python
weather
```

```python
weather['prev_temperature'] = weather.sort_values('recordDate',ascending= True)[['temperature']].shift(1)
weather['prev_date'] = weather.sort_values('recordDate',ascending= True)[['recordDate']].shift(1)
weather['yesterday_date'] = weather['recordDate'] - pd.Timedelta(days=1)
weather[
    (weather.temperature > weather.prev_temperature)
    &
    (weather.prev_date == weather.yesterday_date)
    
][['id']]
```
