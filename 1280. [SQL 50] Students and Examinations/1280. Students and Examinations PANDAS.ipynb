```python
import pandas as pd
```

```python
data = [[1, 'Alice'], [2, 'Bob'], [13, 'John'], [6, 'Alex']]
students = pd.DataFrame(data, columns=['student_id', 'student_name']).astype({'student_id':'Int64', 'student_name':'object'})
data = [['Math'], ['Physics'], ['Programming']]
subjects = pd.DataFrame(data, columns=['subject_name']).astype({'subject_name':'object'})
data = [[1, 'Math'], [1, 'Physics'], [1, 'Programming'], [2, 'Programming'], [1, 'Physics'], [1, 'Math'], [13, 'Math'], [13, 'Programming'], [13, 'Physics'], [2, 'Math'], [1, 'Math']]
examinations = pd.DataFrame(data, columns=['student_id', 'subject_name']).astype({'student_id':'Int64', 'subject_name':'object'})

students
```

```python
stud_sub_df = pd.merge(students,subjects, how = "cross")
stud_sub_df
```

```python
stud_sub_df = pd.merge(students,subjects, how = "cross")

pd.merge(
    stud_sub_df,
    examinations,
    on=['student_id','subject_name'],
    how='left',
    suffixes=('', '_exm')
).groupby(['student_id','student_name','subject_name']).subject_name_exm.count().reset_index().sort_values(
    ['student_id','subject_name']
)
```

```python
stud_sub_df = pd.merge(students, subjects, how="cross")

result = (
    pd.merge(
        stud_sub_df,
        examinations,
        on=['student_id', 'subject_name'],
        how='left',
        suffixes=('', '_exm')
    )
    .groupby(
        ['student_id', 'student_name', 'subject_name'],
        as_index=False
    )
    .agg(attended_exams=('student_id_exm', 'count'))
    .sort_values(['student_id', 'subject_name'])
)

result
```

```python
result = (
    stud_sub_df
    .merge(examinations, on=['student_id', 'subject_name'], how='left')
    .groupby(
        ['student_id', 'student_name', 'subject_name']
    )
    .size()
    .reset_index(name='attended_exams')
    .sort_values(['student_id', 'subject_name'])
)
result
```
